# 아동 특성 관찰 split해서 더미화 -> df_c

```python
# 틱(음성,신체,뚜렛) 을 틱장애 로 변경 - split할때 잘못걸러져서 미리 처리
for i in range(len(Child['아동_특성_관찰'])):
    try:
        Child['아동_특성_관찰'][i] = Child['아동_특성_관찰'][i].replace("틱(음성,신체,뚜렛)","틱장애")
    except AttributeError:
        continue
        
```

```python
ch = []
count =0

for i in range(len(Child['아동_특성_관찰'])):
    try:
        for j in range(len(Child['아동_특성_관찰'][i].split(','))):
            ch.append(Child['아동_특성_관찰'][i].split(',')[j])
    except AttributeError:
        ch.append(np.nan)
    count+=1
print(count)

len(ch)
    
```

```python
ch_c = list(set(ch)) # 리스트 중복 제거
s = pd.Series(ch)    # 시리즈화
s.value_counts()     # 특성별 개수파악
```

```python
s= pd.Series(ch_c)
s.value_counts()
len(ch_c)
# ch_c = ch_c[1:]
ch_c
```

```python
#아동_특성_관찰을 각각 변수로 더미화시키기위해 빈 DataFrame생성 
df_c = pd.DataFrame(0,index=Child.index, columns=ch_c)
df_c
```

```python
#1행 확인
Child['아동_특성_관찰'][0].split(',')
```

```python
#컬럼 확인 
df_c.columns
```

```python
#nan, 파악안됨 컬럼 삭제
df_c = df_c.drop([np.nan,'파악안됨'],axis=1)
df_c
```

```python
#컬럼 다시 확인 
df_c.columns
```

```python
#15분?걸림
#split한 고유의 특성 값이 있으면 DataFrame에 1로 채움 
for i in range(len(Child['아동_특성_관찰'])): # 12만개 한바퀴
    for j in df_c.columns: # 컴럼명과 일치성여부
        try:
            for k in range(len(Child['아동_특성_관찰'][i].split(','))): # 컬럼명과 일치하면 삽입
                if (Child['아동_특성_관찰'][i].split(',')[k]==j):
                    df_c[j][i]=1
                    continue
        except:
            continue
```
