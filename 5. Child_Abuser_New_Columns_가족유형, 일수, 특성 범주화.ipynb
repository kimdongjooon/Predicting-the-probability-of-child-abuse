# 이전신고에서 이후신고까지 걸린 일수 -> New_신고_신고일수
# 이전 조치에서 이후신고까지 걸린 일수 -> New_조치_신고일수
# New_신고_신고일수 / New_조치_신고일수 -> 통계자료기반으로 범주화

### New_신고_신고일수

```python
# 신고-신고
child_lst = sorted(df7['피해아동대상자'].unique()) # 피해아동 리스트 저장
df7['New_신고_신고일수']=np.nan
for n in [2,3]:
    df8 = df7[df7.New_학대횟수==n] # 횟수 지정
    for i in child_lst:
        df9 = df8[df8.피해아동대상자==i]
        for j in range(len(df9.index)-1):
            try:
                t2 = datetime.strptime(df9['신고_접수일시'][df9.index[j+1]], '%Y-%m-%d %H:%M:%S') # 다음행 접수일시
                t3 = datetime.strptime(df9['신고_접수일시'][df9.index[j]], '%Y-%m-%d %H:%M:%S') # 현행 접수일시
                result = t2 - t3 # 다음행 접수일시 - 현행 접수일시 
                if(timedelta(days=0)<=result):
                    df7['New_신고_신고일수'][df9.index[j+1]] = int(str(result).split('day')[0])
            except ValueError:
                if(result<timedelta(days=1)):
                    df7['New_신고_신고일수'][df9.index[j+1]] = 0
                if(result=='0:00:00'):
                    df7['New_신고_신고일수'][df9.index[j+1]] =  '없는게정답'  
    
```

### New_조치_신고일수

```python
# 신고-조치
child_lst = sorted(df7['피해아동대상자'].unique()) # 피해아동 리스트 저장
df7['New_조치_신고일수']=np.nan
for n in [2,3]:
    df8 = df7[df7.New_학대횟수==n] # 횟수 지정
    for i in child_lst:
        df9 = df8[df8.피해아동대상자==i]
        for j in range(len(df9.index)-1):
            t1 = str(int(df9['결과_조치결과일자'][df9.index[j]])) # 조치결과일자
            t1 = datetime.strptime(t1[:4]+'-'+t1[4:6]+'-'+t1[6:8],'%Y-%m-%d')
            t2 = datetime.strptime(df9['신고_접수일시'][df9.index[j+1]], '%Y-%m-%d %H:%M:%S') # 다음행 접수일시
            result = t2 - t1 # 다음행 접수일시 - 현행 조치일시 
            for k in range(len(df9.index)-1):
                try:
                    if(timedelta(days=0)<=result):
                        df7['New_조치_신고일수'][df9.index[k+1]] = int(str(result).split('day')[0])
                except ValueError:
                    if(result<timedelta(days=1)): # 하루 이내
                        df7['New_조치_신고일수'][df9.index[k+1]] = 0
                    if(result=='0:00:00'): # 신고 접수 날짜와 조치결과날짜가 같을때
                        df7['New_조치_신고일수'][df9.index[k+1]] = '이게나라냐'    
```

```python
# 일수 범주화 신고-신고
for i in df7['New_신고_신고일수'].index:
    if(0 <= df7['New_신고_신고일수'][i] <30):
        df7['New_신고_신고일수'][i] = '1개월 이내'
    elif(30 <= df7['New_신고_신고일수'][i] <90):
        df7['New_신고_신고일수'][i] = '1~3 개월 사이'
    elif(90 <= df7['New_신고_신고일수'][i] <180):
        df7['New_신고_신고일수'][i] = '3~6 개월 사이'
    elif(180 <= df7['New_신고_신고일수'][i] <365):
        df7['New_신고_신고일수'][i] = '6개월~1년 사이'
    elif(365 <= df7['New_신고_신고일수'][i] <730):
        df7['New_신고_신고일수'][i] = '1년 ~ 2년 사이'
    elif(730 <= df7['New_신고_신고일수'][i]<1095):
        df7['New_신고_신고일수'][i] = '2년 ~ 3년 사이'
    elif(1095 <= df7['New_신고_신고일수'][i]):
        df7['New_신고_신고일수'][i] = '3년 이상'        
    
```

```python
# 일수 범주화 신고-조치
for i in df7['New_조치_신고일수'].index:
    if(0 <= df7['New_조치_신고일수'][i] <30):
        df7['New_조치_신고일수'][i] = '1개월 이내'
    elif(30 <= df7['New_조치_신고일수'][i] <90):
        df7['New_조치_신고일수'][i] = '1~3 개월 사이'
    elif(90 <= df7['New_조치_신고일수'][i] <180):
        df7['New_조치_신고일수'][i] = '3~6 개월 사이'
    elif(180 <= df7['New_조치_신고일수'][i] <365):
        df7['New_조치_신고일수'][i] = '6개월~1년 사이'
    elif(365 <= df7['New_조치_신고일수'][i] <730):
        df7['New_조치_신고일수'][i] = '1년 ~ 2년 사이'
    elif(730 <= df7['New_조치_신고일수'][i]<1095):
        df7['New_조치_신고일수'][i] = '2년 ~ 3년 사이'
    elif(1095 <= df7['New_조치_신고일수'][i]):
        df7['New_조치_신고일수'][i] = '3년 이상'  
        
```
