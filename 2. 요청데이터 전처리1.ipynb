# 결측치 확인 및 처리

```python
pd.set_option('display.max_rows',95)
# 총 93개 변수에 대해서 결측치 수와 비율 확인
Child_missing_df = Child.isnull().sum().reset_index()
Child_missing_df.columns = ['column', 'count']
Child_missing_df['ratio'] = round(Child_missing_df['count'] / Child.shape[0]*100,4)
Child_missing_df
```

```python
#결측치 99%이상인 컬럼 추출
Child_missing99=Child_missing_df.loc[Child_missing_df['ratio'] >=99]      
Child_missing99.reset_index(inplace = True)
Child_missing99.drop(['index'],axis=1)
```

# 아동_생년월일 -> 연령대로 변환(신고_접수일시 이용)

```python
int(Child['아동_생년월일'][0][:2]) # 생년월일
# Child['Unnamed: 4'][5543]
```

```python
# 접수일시에서의 연도에서 생년월일의 생년을 빼서 나온수치를 아동의 나이로 사용할 예정
for i in range(len(Child)):
    try:
        age = int(str(Child['신고_접수일시'][i])[2:4]) - int(Child['아동_생년월일'][i][:2]) 
        if (age < 0):
            age = 100 + int(str(Child['신고_접수일시'][i])[1:4]) -  int(Child['아동_생년월일'][i][:2])
        if ((0<= age <=17) != True):
            age = np.nan
        Child['아동_생년월일'][i] = age
    except TypeError:
        Child['아동_생년월일'][i] = np.nan
    except ValueError:
        Child['아동_생년월일'][i] = np.nan
```

```python
print(Child['아동_생년월일'].value_counts())
print("null 값: ",Child['아동_생년월일'].isnull().sum())
```

```python
#통계 자료 기준으로 연령대 범주화
for i in range(len(Child['아동_생년월일'])):
    try:
        if (0<= Child['아동_생년월일'][i] < 1 ):
            Child['아동_생년월일'][i] = '1세미만'
        elif (1<= Child['아동_생년월일'][i] <= 3):
            Child['아동_생년월일'][i] = '1~3세'
        elif (4<= Child['아동_생년월일'][i] <= 6):
            Child['아동_생년월일'][i] = '4~6세'
        elif (7<= Child['아동_생년월일'][i] <= 9):
            Child['아동_생년월일'][i] = '7~9세'
        elif (10<= Child['아동_생년월일'][i] <= 12):
            Child['아동_생년월일'][i] = '10~12세'
        elif (13<= Child['아동_생년월일'][i] <= 15):
            Child['아동_생년월일'][i] = '13~15세'
        elif (16<= Child['아동_생년월일'][i] <= 17):
            Child['아동_생년월일'][i] = '16~17세'
    except ValueError:
        continue
```

```python
print(Child['아동_생년월일'].value_counts())
print("null 값: ",Child['아동_생년월일'].isnull().sum() )
```
