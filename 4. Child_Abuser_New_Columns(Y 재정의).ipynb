```python
#122042 rows × 256 columns
Child_Abuser = pd.read_csv('./Child_Abuser.csv',encoding = 'CP949')
Child_Abuser
```

```python
df1_c = Child_Abuser
df1 = Child_Abuser
```

# New_학대횟수

```python
df1 = df1[df1['판단_학대혐의여부']=='Y']
df1.columns
for i in df1:
    print(i)
```

```python
# Y값 설정에 필요한 컬럼 추출
df2 = df1[['피해아동대상자','신고_접수일시','결과_조치결과일자','판단_학대혐의여부','결과_행위자조치결과','결과_아동조치결과']]
# 결측지 제거 122042 -> 87487
df2 = df2.dropna(axis=0,subset=['결과_조치결과일자','판단_학대혐의여부'])
df2 = df2.drop(['판단_학대혐의여부'],axis=1)
df2
```

```python
df2index = df2[df2.duplicated(['피해아동대상자','신고_접수일시'], keep=False)==True].index
df2index
```

# 다중학대행위자: 한 피해아동의 행위자가 여러명인 경우

```python
df2['다중학대행위자']=np.nan
df2
```

```python
df2.loc[df2index,'다중학대행위자']='Y'
```

```python
# 중복 삭제
df2_d = df2.drop_duplicates(['피해아동대상자','신고_접수일시'], keep='first')
```

```python
#확인
df2_d[df2_d.duplicated(['피해아동대상자','신고_접수일시'], keep='first') ==True]
```

```python
# df2['New_학대횟수']=0
df2_d['New_학대횟수']=0
```

```python
# df3 = df2.sort_values(by=["피해아동대상자", "신고_접수일시",'결과_조치결과일자'], ascending=[True,True,True])
df3 = df2_d.sort_values(by=["피해아동대상자", '신고_접수일시','결과_조치결과일자'], ascending=[True,True,True])
df3[:50]
```

```python
count=1 # 학대횟수 
idx_dup_lst=[] # 중복행 제거 리스트
idx_del_lst=[] # 비해당행 제거 리스트
child_lst1 = sorted(df3['피해아동대상자'].unique()) # 피해아동 리스트 저장
for i in child_lst1:
    df4 = df3[df3['피해아동대상자']==i] # 피해아동 검색후
    if(len(df4)>1):
        for j in range(len(df4.index)-1):
            t2 = datetime.strptime(df4['신고_접수일시'][df4.index[j+1]], '%Y-%m-%d %H:%M:%S') # 다음행 접수일시
            t3 = datetime.strptime(df4['신고_접수일시'][df4.index[j]], '%Y-%m-%d %H:%M:%S') # 현행 접수일시
            result2 = t2 - t3 # 다음행 접수일시 - 현행 접수일시 
            if(timedelta(days=0)<=result2 < timedelta(days=2)):
                df4_d = df4.loc[[df4.index[j],df4.index[j+1]],:] # 두행의 중복성을 알아보기위한 데이터프레임 지정
                if(df4_d.duplicated(['결과_조치결과일자','결과_아동조치결과','결과_행위자조치결과'])[df4.index[j+1]] == True):
                    idx_dup_lst.append(df4.index[j+1]) # 중복행 제거 리스트에 저장
df5 = df3.drop(idx_dup_lst)
child_lst2 = sorted(df5['피해아동대상자'].unique()) # 피해아동 리스트 저장
for i in child_lst2:
    df6 = df5[df5['피해아동대상자']==i] # 피해아동 검색후
    if(len(df6)>1):
        for j in range(len(df6.index)-1):
            t1 = str(int(df6['결과_조치결과일자'][df6.index[0]])) # 조치결과일자
            t1 = datetime.strptime(t1[:4]+'-'+t1[4:6]+'-'+t1[6:8],'%Y-%m-%d')
            t2 = datetime.strptime(df6['신고_접수일시'][df6.index[j+1]], '%Y-%m-%d %H:%M:%S') # 다음행 접수일시
            result1 = t2 - t1 # 다음행 접수일시 - 현행 조치결과 일시
            if(timedelta(days=0)<result1):
                count+=1
            if(result1<=timedelta(days=0)):# 조치 결과일자보다 신고접수가 먼저 들어왔을때 
                idx_del_lst.append(df6.index[j+1]) # 비해당행 제거 리스트에 저장
    for k in df6.index:
        df5['New_학대횟수'][k]=count
    count=1
df7 = df5.drop(idx_del_lst)
df7
```

# '재학대횟수'로 'New_재학대혐의_수정' 생성

```python
df7['New_재학대혐의_수정']=np.nan
df7
```

```python
for i in df7.index:
    if(df7['New_학대횟수'][i]<=1):
        df7['New_재학대혐의_수정'][i]='N'
    elif(df7['New_학대횟수'][i]>1):
        df7['New_재학대혐의_수정'][i]='Y'
```

```python
df7[:60]
```

```python
df7['New_재학대혐의_수정'].value_counts()
```

```python
df1_c['다중학대행위자'] = np.nan
df1_c['New_학대횟수'] = 0
df1_c['New_재학대혐의_수정'] =np.nan
for i in df1_c.index:
    for j in df7.index:
        if (i == j):
            df1_c['다중학대행위자'][i] = df7['다중학대행위자'][j]
            df1_c['New_학대횟수'][i] = df7['New_학대횟수'][j]
            df1_c['New_재학대혐의_수정'][i] = df7['New_재학대혐의_수정'][j]
```

```python
df = df1_c[['피해아동대상자','신고_접수일시','결과_조치결과일자','결과_아동조치결과','결과_행위자조치결과','판단_학대혐의여부','New_학대횟수','New_재학대혐의_수정']]
df = df.sort_values(by=['피해아동대상자','신고_접수일시','결과_조치결과일자'],ascending=[True,True,True])
df
```
